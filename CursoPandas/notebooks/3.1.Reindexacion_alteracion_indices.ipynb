```python
import pandas as pd
```

```python
data = {
    'Producto':['Manzanas', 'Naranjas', 'Platanos', 'Uvas', 'Peras'],
    'Precio': [100, 80, 60, 120, 90]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.set_index('Producto', inplace = True)
```

```python
print(df)
```

```python
nuevo_orden = ['Uvas', 'Manzanas', 'Melones', 'Naranjas', 'Peras','Platanos']
df_reindexado = df.reindex(nuevo_orden)
print(df_reindexado)
```

```python
df_renombrado = df.rename(index ={'Manzanas': 'Apples', 'Naranjas': 'Oranges'})
print(df_renombrado)
```

```python
df_reset = df.reset_index()
print(df_reset)
```
