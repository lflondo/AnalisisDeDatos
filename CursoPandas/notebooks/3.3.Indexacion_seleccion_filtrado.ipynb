```python
import pandas as pd
```

```python
data = {
    'Producto':['Manzanas', 'Naranjas', 'Platanos', 'Uvas', 'Peras'],
    'Precio': [100, 80, 60, 120, 90],
    'Stock':[30,50,20,60,40]
}

df = pd.DataFrame(data)
df
```

## Indexados

```python
precios = df['Precio']
print(precios)
```

#### Seleccion por Etiquetas o nombres de filas Loc

```python
producto_naranjas = df.loc[1,'Producto']
print(producto_naranjas)
```

```python
producto_naranjas = df.loc[1]
print(producto_naranjas)
```

#### Seleccion por posicion Iloc

```python
producto_posicion = df.iloc[2,1]
print(producto_posicion)
```

```python
producto_posicion = df.iloc[2,0]
print(producto_posicion)
```

```python
df_nuevo = df.set_index("Producto")
df_nuevo
```

```python
df_nuevo.loc["Peras"]
```

## Filtrados

```python
df_precios_altos = df[df['Precio']>80]
df_precios_altos
```

```python
df_filtrado =df[(df['Precio'] > 60) & (df['Stock']>30)]
```

```python
df_filtrado
```

```python
df_where = df['Precio'].where(df['Precio']>80, other=0)
df_where
```

```python
df_query = df.query('Precio > 80')
df_query
```
