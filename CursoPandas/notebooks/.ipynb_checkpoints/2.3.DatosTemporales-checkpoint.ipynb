```python
import pandas as pd
```

```python
fechas = pd.date_range(start = '2020-01-01', periods=6, freq='D')
valores = [34, 44, 65, 53, 39, 76]
serie_temporal = pd.Series(valores, index=fechas)
print(serie_temporal)
```

```python
print(serie_temporal ['2020-01-06'])
```

```python
print(serie_temporal['2020-01-02':'2020-01-05'])
```

```python
data ={
    'Fecha': pd.date_range(start='2020-01-01', periods=6, freq='D'),
    'Ventas': [345, 456, 567, 678, 789, 890]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.set_index('Fecha', inplace=True)
print(df)
```

```python
resumen_mensual = df.resample ('ME').mean() ### Se utiliza ('YE') para a√±o, ('ME') por mes y ('D') por dia (h) hora, (min) minutos, ('s') segundos
print(resumen_mensual)
```

```python
df['Ventas_desplazadas'] = df['Ventas'].shift(1) ### shift desplaza hacia delante el va que se le ponga en este caso 1
df['Cambio_ventas'] = df['Ventas'].diff() ### diff calcula la diferencia entre cada punto y su anterior.
print(df)
```
