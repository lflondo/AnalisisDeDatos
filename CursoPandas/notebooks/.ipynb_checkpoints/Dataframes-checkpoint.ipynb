```python
import pandas as pd
```

```python
data = [
        {'Nombre': 'Camilo', 'Edad':25, 'Ciudad': 'Bucaramanga'},
        {'Nombre': 'Roman', 'Edad':16, 'Ciudad': 'Bogotá'},
        {'Nombre': 'Andrea', 'Edad':32, 'Ciudad': 'Cali'},
        {'Nombre': 'Luis', 'Edad':44, 'Ciudad': 'Medellín'}
       ]
df = pd.DataFrame(data)
df
```

# DataFrame formado desde lista

```python
datalist = {
    'Nombre': ['Camilo', 'Ana', 'Pedro', 'Marta'],
    'Edad': [25,16,34,12],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']
}
df = pd.DataFrame(datalist)
df
```

#### Seleccionar columnas del Dataframe

```python
nombres = df['Nombre']
nombres
```

#### Añadir columna

```python
df['Salario'] = [30000, 40000, 25000, 32000]
df
```

#### Borrar columna

```python
df.drop('Salario', axis=1, inplace=True)
df
```

#### Seleccionar filas por posición

```python
primera_fila = df.iloc[3]
primera_fila
```

####  Seleccionar filas por condiciones

```python
datadux = {
    'Nombre': ['Camilo', 'Ana', 'Pedro', 'Marta', 'Ana'],
    'Edad': [25,16,None,12,56],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', None]
}
df = pd.DataFrame(datadux)
df
```

```python
df.loc[df['Nombre'] =='Ana']
```

```python
df[df['Edad']<30]
```

#### Ordenar el Dataframe

```python
df.sort_values(by='Edad', ascending=True)
```

```python
df['Edad'].mean()
```

```python
df['Edad'].max()
```

```python
df_limpio =df.dropna() #Elimina filas con elementos vacios, no altera el dataframe sino que crea uno nuevo sin estos elementos.
df_limpio
```
