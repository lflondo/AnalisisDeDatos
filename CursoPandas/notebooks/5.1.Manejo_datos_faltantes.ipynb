```python
import pandas as pd
```

```python
data ={'A': [1, 2, None, 4, 5],
       'B':[None,4,3,2,None,],
       'C':[2,3,4,None,6]}
df = pd.DataFrame(data)
df
```

```python
faltantes_por_columna = df.isnull().sum()
print("Número de valores faltantes por columna:\n", faltantes_por_columna)
```

```python
df_sin_filas_faltantes = df.dropna()
print("DartaFrame sin filas con valores faltantes:\n", df_sin_filas_faltantes)
```

```python
df_sin_filas_faltantes = df.dropna(axis = 1)
print("DartaFrame sin filas con valores faltantes:\n", df_sin_filas_faltantes)
```

```python
df_datos_relenados = df.fillna(0)
print("DartaFrame con valores faltantes rellenados con 0:\n", df_datos_relenados)
```

```python
df_datos_relenados = df.fillna(df.mean())
print("DartaFrame con valores faltantes rellenados con la media:\n", df_datos_relenados)
```

```python
df_rellenado_ffill = df.ffill()
print("DartaFrame con valores faltantes rellenados con el valor anterior:\n", df_rellenado_ffill)
```

```python
df_rellenado_bfill = df.bfill()
print("DartaFrame con valores faltantes rellenados con el valor siguiente:\n", df_rellenado_bfill)
```
