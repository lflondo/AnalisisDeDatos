```python
angles = np.array([0,30,60,90])
```

```python
np.sin(angles)
```

```python
angles_rad = np.radians(angles)
```

```python
np.sin(angles_rad)
```

```python
np.cos(angles_rad)
```

```python
np.tan(angles_rad)
```
