```python
m = np.array([1,5,7])
n= m.copy() #Esto consume mas memoria, pero si necesitamos crear independiancias entre los array se debe utilizar copy.

m[1] = 9
```

```python
print('m: ',m)
print('n: ',n)
```

```python
sumar = np.sum(b)
print('Suma del array b:', sumar)
minimo = np.min(b)
print('numero minimo del array b:', minimo)
maximo = np.max(b)
print('numero maximo del array b:', maximo)
media = np.mean(b)
print('media del array b:', media)
desv_standar = np.std(b)
print('desviacion estandar del array b:', desv_standar)
######-----------------------------------------------------------
print()
media_row = np.mean(b, axis = 0) #Media de las filas
print('media de las filas del array b:', media_row)
print()
min_colum = np.min(b, axis = 1) #minimo de las columnas
print('minimo de las columnas del array b:', min_colum)
```
