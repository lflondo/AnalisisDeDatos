```python
import numpy as np
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[2,0],[1,2]])
```

```python
# Multiplicar matrices forma 1:
C = np.dot(A,B)
print(C)
```

```python
# Multiplicar matrices forma 2:
D = A @ B
print(D)
```

```python
np.linalg.det(A) #Obtener el determinante
```

```python
#Calcular la inversa 
det_A = np.linalg.det(A)
if det_A !=0:
    inv_A = np.linalg.inv(A)
    print("Inversa de A: \n", inv_A)
else:
    print("A no es invertible")
```

```python
matrix_M = np.array([[3,3],[3,3]])
```

```python
#Calcular la inversa 
det_M = np.linalg.det(matrix_M)
print("Determinante de M:",det_M)
print()
if det_M !=0:
    inv_M = np.linalg.inv(matrix_M)
    print("Inversa de M: \n", inv_M)
else:
    print("M no es invertible, por que su determinante es 0")
```

```python
#Descomposiciones de matrices
print("Matrix A:\n",A)
print()
Q, R = np.linalg.qr(A)

print("Factor Q:\n",Q)
print()
print("Factor R:\n",R)
```
