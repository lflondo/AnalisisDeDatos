```python
import numpy as np
```

```python
A = np.array ([[3,1], [1,2]])
print("Matriz A:\n",A)
b = np.array([9, 8])
print("Vector b:\n",b)
```

```python
x= np.linalg.solve(A, b)
print("La solución del sistema:\n", x)
```

```python
print("Verificacion de Ax = b:\n", np.dot(A,x))
```

```python
A = np.array ([[1,2], [3,4],[5,6]])
print("Matricial A:\n",A)
b = np.array([1,2,3])
print("Vector b:\n",b)
```

```python
A_pinv = np.linalg.pinv(A)
print(A_pinv)
```

```python
x = np.dot(A_pinv, b)
print(x)
```
