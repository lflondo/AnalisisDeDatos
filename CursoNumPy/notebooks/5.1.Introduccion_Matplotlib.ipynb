```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

Graficar: $f(x) = x^2-4x+4$

```python
x = np.linspace(-1,5.400)
```

```python
y = x**2 - 4*x + 4
```

```python
plt.figure(figsize =(8, 4))
plt.plot(x, y, label = 'f(x) = x^2-4x+4', color='red')
plt.title('Grafico de Dispersión Básico')
plt.xlabel('X')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
