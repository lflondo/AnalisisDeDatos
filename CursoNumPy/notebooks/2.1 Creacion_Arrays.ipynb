```python
##Array de ceros y unos

zeros_array = np.zeros ((2,3))
ones_array = np.ones((3,2))

print ("Array de ceros:\n", zeros_array)
print ("Array de unos:\n", ones_array)
```

```python
## Arrays de rangos
range_array = np.arange(10)
print ("Range Array: ", range_array)
```

```python
## Array espacioados liealmente

linspace_array = np.linspace(-1, 2, num=5)

print("linspace_array: ", linspace_array)
```

```python
## Matriz identidad

identity_matrix = np.eye(5,6)
print("Matriz de identidad: ", identity_matrix)
```

```python
## Array con tipos de datos especificos
int_array = np.array([1,2,3], dtype='int32')
float_array = np.array([1,2,3], dtype='float32')
print("Integer Array: ", int_array)
print("Float Array: ", float_array)
```
